# 2. عمليات DataFrame الأساسية

```python
from pyspark.sql import SparkSessionspark = SparkSession.builder.appName('DataFrameOperations').getOrCreate()df = spark.read.csv('../data/yellow_tripdata_2016-03.csv', header=True, inferSchema=True)
```

## اختيار الأعمدة

```python
df.select(['tpep_pickup_datetime', 'trip_distance', 'total_amount']).show(5)
```

## إضافة عمود جديد

```python
from pyspark.sql.functions import col, litdf_with_new_col = df.withColumn('trip_distance_km', col('trip_distance') * 1.60934)df_with_new_col.select(['trip_distance', 'trip_distance_km']).show(5)
```

## حذف عمود

```python
df_dropped = df_with_new_col.drop('trip_distance_km')print(df_dropped.columns)
```

## إعادة تسمية عمود

```python
df.withColumnRenamed('total_amount', 'total_fare').show(5)
```
